```r
install.packages("ISLR")
install.packages("lubridate")
install.packages("data.table")
install.packages("DMwR2")
install.packages("ggplot2")
install.packages("dplyr")
library(ggplot2)
library(DMwR2)
library(ISLR)
library(lubridate)
library(data.table)
library(dplyr)
##################### Task 1 #####################

BetsData = read.csv(file="C:/Users/Ekin/Desktop/Ders/0.Master/IE582/Homework 1 Yeniden/bets.csv", header=TRUE, sep=",")
BookingData = read.csv(file="C:/Users/Ekin/Desktop/Ders/0.Master/IE582/Homework 1 Yeniden/booking.csv", header=TRUE, sep=",")
GoalsData = read.csv(file="C:/Users/Ekin/Desktop/Ders/0.Master/IE582/Homework 1 Yeniden/goals.csv", header=TRUE, sep=",")
MatchesData = read.csv(file="C:/Users/Ekin/Desktop/Ders/0.Master/IE582/Homework 1 Yeniden/matches.csv", header=TRUE, sep=",")
StatsData = read.csv(file="C:/Users/Ekin/Desktop/Ders/0.Master/IE582/Homework 1 Yeniden/stats.csv", header=TRUE, sep=",")
PremierMatchesData = subset(MatchesData, league_id=='148')

#Omit NA values from the data
HomeX = PremierMatchesData$match_hometeam_score
HomeX2=na.omit(HomeX)
AwayX = PremierMatchesData$match_awayteam_score
AwayX2=na.omit(AwayX)
HomeAwayX = PremierMatchesData$match_hometeam_score - PremierMatchesData$match_awayteam_score
HomeAwayX2=na.omit(HomeAwayX)

#Draw the histograms
hist1 = hist(HomeX2,breaks=8,col="blue", main="Histogram of Home Goals",xlab="Home Goals", ylab="Number of Games")
hist2 = hist(AwayX2,breaks=6,col="red",main="Histogram of Away Goals",xlab="Away Goals", ylab="Number of Games")
hist3 = hist(HomeAwayX2,breaks=14,col="green", main="Histogram of Home Goals - Away Goals",xlab="Home Goals - Away Goals", ylab="Number of Games")

#Home and away goals look like Poisson distribution, whereas Home-Away goals looks like normal distribution.
#Generate subsets with Poisson distribution
PoissonHome = rpois(618,mean(HomeX2))
PoissonAway = rpois(618,mean(AwayX2))

max(HomeX2)
#Count the numbers in theoretical distribution
PoissonHome0 = dpois(0,mean(HomeX2))*618
PoissonHome1 = dpois(1,mean(HomeX2))*618
PoissonHome2 = dpois(2,mean(HomeX2))*618
PoissonHome3 = dpois(3,mean(HomeX2))*618
PoissonHome4 = dpois(4,mean(HomeX2))*618
PoissonHome5 = dpois(5,mean(HomeX2))*618
PoissonHome6 = dpois(6,mean(HomeX2))*618
PoissonHome7 = dpois(7,mean(HomeX2))*618
PoissonHome8 = dpois(8,mean(HomeX2))*618
#Checking the assignment is true or not
PoissonHome0 + PoissonHome1 + PoissonHome2 + PoissonHome3 + PoissonHome4 + PoissonHome5 + PoissonHome6 + PoissonHome7 + PoissonHome8 

max(AwayX2)
#Count the numbers in theoretical distribution
PoissonAway0 = dpois(0,mean(AwayX2))*618
PoissonAway1 = dpois(1,mean(AwayX2))*618
PoissonAway2 = dpois(2,mean(AwayX2))*618
PoissonAway3 = dpois(3,mean(AwayX2))*618
PoissonAway4 = dpois(4,mean(AwayX2))*618
PoissonAway5 = dpois(5,mean(AwayX2))*618
PoissonAway6 = dpois(6,mean(AwayX2))*618

#Plot histograms and distributions

hist1 = hist(HomeX2,breaks=8,col="blue", main="Histogram of Home Goals",xlab="Home Goals", ylab="Number of Games")
xfit=seq(from=0,to=8)
yfit = c(PoissonHome0,PoissonHome1,PoissonHome2,PoissonHome3,PoissonHome4,PoissonHome5,PoissonHome6,PoissonHome7,PoissonHome8)
lines(xfit, yfit, col="red", lwd=2)

hist2 = hist(AwayX2,breaks=8,col="red", main="Histogram of Away Goals",xlab="Away Goals", ylab="Number of Games")
xfit1=seq(from=0,to=6)
yfit1 = c(PoissonAway0,PoissonAway1,PoissonAway2,PoissonAway3,PoissonAway4,PoissonAway5,PoissonAway6)
lines(xfit1, yfit1, col="blue", lwd=2)

hist3 = hist(HomeAwayX2,breaks=14,col="green", main="Histogram of Home Goals - Away Goals",xlab="Home Goals - Away Goals", ylab="Number of Games")
xfit2 = seq(min(HomeAwayX2),max(HomeAwayX2),length=618)
yfit2 = dnorm(xfit2,mean=mean(HomeAwayX2),sd=sd(HomeAwayX2))
yfit2 = yfit2*diff(hist3$mids[1:2])*length(HomeAwayX2)
lines(xfit2, yfit2, col="black", lwd=2)

mean(HomeX2)
mean(AwayX2)
var(HomeX2)
var(AwayX2)

#Comments on Task 1
#Both home and away goals distributed approximately like Poisson. Home-Away goals are distributed like Normal distribution, and the mean is greater than zero.
#It is obvious that home teams are scoring more than away goals, but with a higher variance. 
#It is possible to score more goals in home games, both in average goals and maximum goals.
#In fact, away goals seem to be much higher than I was expecting at the beginning of this assignment. 
#Difference between home and away goals is distributed normally with respect to central limit theorem. However, histogram seems to shift towards negative side.
#In other words, the distribution is not symmetric in home goals - away goals. If away team is winning the game, it is possible for the away team to score more than expected. (?)
#Lastly, both home and away teams seem to score 0-1 goals more than actual poisson distribution would create. Scoring goals seems to be harder than expectation in both sides.
#Scoring more than 1 goal is occured less than expected in both home and away teams. 

##################### Task 2 #####################

#####1xBet BET#####
PremierMatches = filter(MatchesData, MatchesData$league_id == "148")

BetsData1 = filter(BetsData, BetsData$odd_bookmakers == "1xBet" & BetsData$match_id %in% PremierMatches$match_id )
BetsData1Home = filter(BetsData1, BetsData1$variable == "odd_1")
BetsData1Away = filter(BetsData1, BetsData1$variable == "odd_2")
BetsData1Draw = filter(BetsData1, BetsData1$variable == "odd_x")

###Part 1 Calculating the probabilities
PHomeWin1 = 1/BetsData1Home$value
PAwayWin1 = 1/BetsData1Away$value
PDraw1 = 1/BetsData1Draw$value

###Part 2 Normalization of probabilities
PHomeWin1Normal = PHomeWin1 / (PHomeWin1 + PAwayWin1 + PDraw1)
PAwayWin1Normal = PAwayWin1 / (PHomeWin1 + PAwayWin1 + PDraw1)
PDraw1Normal = PDraw1 / (PHomeWin1 + PAwayWin1 + PDraw1)

###Part 3
DrawMatches = filter(MatchesData, MatchesData$match_hometeam_score - MatchesData$match_awayteam_score == 0 & MatchesData$league_id == "148")
plot(PHomeWin1-PAwayWin1, PDraw1,xlab= "P(Home)-P(Away)", ylab = "P(Draw)",col="red",main="1xBet")
plot(PHomeWin1Normal-PAwayWin1Normal,PDraw1Normal,xlab= "P(Home)-P(Away) Normalized", ylab = "P(Draw) Normalized",col="blue",main="1xBet")
TotalCountBin1=list()
CountBin1 = list()
MyProbability1 = list()
Bias1 = list()
for(i in 1:10){
  TotalCountBin1[i] = sum((PHomeWin1-PAwayWin1)>=(-1.2+0.2*i) & (PHomeWin1-PAwayWin1)<(-1.2+0.2*(i+1)))
}

BetsData1ActualDraw = filter(BetsData1, BetsData1$match_id %in% DrawMatches$match_id)

BetsData1ActualDrawHome = filter(BetsData1ActualDraw, BetsData1ActualDraw$variable == "odd_1")

BetsData1ActualDrawAway = filter(BetsData1ActualDraw, BetsData1ActualDraw$variable == "odd_2")

PHomeWinActual1 = 1/BetsData1ActualDrawHome$value
PAwayWinActual1 = 1/BetsData1ActualDrawAway$value

for(i in 1:10){
  CountBin1[i] = sum((PHomeWinActual1-PAwayWinActual1)>=(-1.2+0.2*i) & (PHomeWinActual1-PAwayWinActual1)<(-1.2+0.2*(i+1)))
}

for(i in 1:10){
  MyProbability1[i] = CountBin1[[i]]/TotalCountBin1[[i]]
}

for(i in 1:10){
  Bias1[i] = sum((PHomeWin1-PAwayWin1)>(-1.2+0.2*i) & (PHomeWin1-PAwayWin1)<=(-1.2+0.2*(i+1)) & MyProbability1[i]>PDraw1 )
}
BiasArray1 = unlist(Bias1)
plot(x=1:10, y=BiasArray1,xlab="Index of the bin",ylab="Number of Biased Odds",main="1xBet Bias Graph",xaxt = "n",yaxt="n")
axis(1, at = seq(1, 10, by = 1), las=2)
axis(2, at=seq(0,max(BiasArray1)+1,by=5), las=2)

##If probability of two teams is similar, it is expected to have bias in the draw bet.

##################### Task 3 #####################

BookingDataWithoutRed = filter(BookingData,BookingData$card=="red card")
##Eliminate matches with a red card in first 15 minutes
BookingDataWithout15 = filter(BookingDataWithoutRed, as.numeric(BookingDataWithoutRed$time) < 15)
BookingDataWithout15Red = filter(BookingData, !(BookingData$match_id %in% BookingDataWithout15$match_id))
##Either a match is going draw, and ended as draw. Or a match is being win by home, ended as home. Or a match is being win by away, ended as away. This is the perfect mathematics for this question. Other approaches might be hold, but there are too many possibilities, be careful!
MatchesDataWithoutSurprise = filter(MatchesData, abs(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score) + abs(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score - MatchesData$match_hometeam_extra_score + MatchesData$match_awayteam_extra_score) == 0 | (MatchesData$match_hometeam_extra_score - MatchesData$match_awayteam_extra_score)*(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score - MatchesData$match_hometeam_extra_score + MatchesData$match_awayteam_extra_score) > 0)

BetsDataWithoutSurprise1 = filter(BetsData1, BetsData1$match_id %in% MatchesDataWithoutSurprise$match_id)
BetsDataWithoutSurprise11 = filter(BetsDataWithoutSurprise1, BetsDataWithoutSurprise1$match_id %in% BookingDataWithout15Red$match_id)                                 

BetsData1HomeWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_1")
BetsData1AwayWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_2")
BetsData1DrawWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_x")

PHomeWin14 = 1/BetsData1HomeWithoutSurprise$value
PAwayWin14 = 1/BetsData1AwayWithoutSurprise$value
PDraw14 = 1/BetsData1DrawWithoutSurprise$value

plot(PHomeWin14-PAwayWin14, PDraw14,xlab= "P(Home)-P(Away)", ylab = "P(Draw)",col="red",main="1xBet without Surprise")
TotalCountBin14=list()
CountBin14 = list()
MyProbability14 = list()
Bias14 = list()

for(i in 1:10){
  TotalCountBin14[i] = sum((PHomeWin14-PAwayWin14)>=(-1.2+0.2*i) & (PHomeWin14-PAwayWin14)<(-1.2+0.2*(i+1)))
}

BetsData1ActualDraw4 = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$match_id %in% DrawMatches$match_id)

BetsData1ActualDrawHome4 = filter(BetsData1ActualDraw4, BetsData1ActualDraw4$variable == "odd_1")
BetsData1ActualDrawAway4 = filter(BetsData1ActualDraw4, BetsData1ActualDraw4$variable == "odd_2")

PHomeWinActual14 = 1/BetsData1ActualDrawHome4$value
PAwayWinActual14 = 1/BetsData1ActualDrawAway4$value

for(i in 1:10){
  CountBin14[i] = sum((PHomeWinActual14-PAwayWinActual14)>=(-1.2+0.2*i) & (PHomeWinActual14-PAwayWinActual14)<(-1.2+0.2*(i+1)))
}

for(i in 1:10){
  MyProbability14[i] = CountBin14[[i]]/TotalCountBin14[[i]]
}

for(i in 1:10){
  Bias14[i] = sum((PHomeWin14-PAwayWin14)>(-1.2+0.2*i) & (PHomeWin14-PAwayWin14)<=(-1.2+0.2*(i+1)) & MyProbability14[i]>PDraw14 )
}
BiasArray14 = unlist(Bias14)
plot(x=1:10, y=BiasArray14,xlab="Index of the bin",ylab="Number of Biased Odds",main="1xBet Without Surprise Bias Graph",xaxt = "n",yaxt="n")
axis(1, at = seq(1, 10, by = 1), las=2)
axis(2, at=seq(0,max(BiasArray1)+1,by=2), las=2)


## Same result holds, if two teams are close to each other by means of winning probability, then a bias is likely to occur in betting draw.
## You can see below other 3 results of the bookmakers

#####BetVictor#####


PremierMatches = filter(MatchesData, MatchesData$league_id == "148")

BetsData1 = filter(BetsData, BetsData$odd_bookmakers == "BetVictor" & BetsData$match_id %in% PremierMatches$match_id )
BetsData1Home = filter(BetsData1, BetsData1$variable == "odd_1")
BetsData1Away = filter(BetsData1, BetsData1$variable == "odd_2")
BetsData1Draw = filter(BetsData1, BetsData1$variable == "odd_x")

###Part 1 Calculating the probabilities
PHomeWin1 = 1/BetsData1Home$value
PAwayWin1 = 1/BetsData1Away$value
PDraw1 = 1/BetsData1Draw$value

###Part 2 Normalization of probabilities
PHomeWin1Normal = PHomeWin1 / (PHomeWin1 + PAwayWin1 + PDraw1)
PAwayWin1Normal = PAwayWin1 / (PHomeWin1 + PAwayWin1 + PDraw1)
PDraw1Normal = PDraw1 / (PHomeWin1 + PAwayWin1 + PDraw1)

###Part 3
DrawMatches = filter(MatchesData, MatchesData$match_hometeam_score - MatchesData$match_awayteam_score == 0 & MatchesData$league_id == "148")
plot(PHomeWin1-PAwayWin1, PDraw1,xlab= "P(Home)-P(Away)", ylab = "P(Draw)",col="red",main="BetVictor")
plot(PHomeWin1Normal-PAwayWin1Normal,PDraw1Normal,xlab= "P(Home)-P(Away) Normalized", ylab = "P(Draw) Normalized",col="blue",main="BetVictor")
TotalCountBin1=list()
CountBin1 = list()
MyProbability1 = list()
Bias1 = list()
for(i in 1:10){
  TotalCountBin1[i] = sum((PHomeWin1-PAwayWin1)>=(-1.2+0.2*i) & (PHomeWin1-PAwayWin1)<(-1.2+0.2*(i+1)))
}

BetsData1ActualDraw = filter(BetsData1, BetsData1$match_id %in% DrawMatches$match_id)

BetsData1ActualDrawHome = filter(BetsData1ActualDraw, BetsData1ActualDraw$variable == "odd_1")

BetsData1ActualDrawAway = filter(BetsData1ActualDraw, BetsData1ActualDraw$variable == "odd_2")

PHomeWinActual1 = 1/BetsData1ActualDrawHome$value
PAwayWinActual1 = 1/BetsData1ActualDrawAway$value

for(i in 1:10){
  CountBin1[i] = sum((PHomeWinActual1-PAwayWinActual1)>=(-1.2+0.2*i) & (PHomeWinActual1-PAwayWinActual1)<(-1.2+0.2*(i+1)))
}

for(i in 1:10){
  MyProbability1[i] = CountBin1[[i]]/TotalCountBin1[[i]]
}

for(i in 1:10){
  Bias1[i] = sum((PHomeWin1-PAwayWin1)>(-1.2+0.2*i) & (PHomeWin1-PAwayWin1)<=(-1.2+0.2*(i+1)) & MyProbability1[i]>PDraw1 )
}
BiasArray1 = unlist(Bias1)
plot(x=1:10, y=BiasArray1,xlab="Index of the bin",ylab="Number of Biased Odds",main="BetVictor Bias Graph",xaxt = "n",yaxt="n")
axis(1, at = seq(1, 10, by = 1), las=2)
axis(2, at=seq(0,max(BiasArray1)+1,by=5), las=2)

##If probability of two teams is similar, it is expected to have bias in the draw bet.

##################### Task 3 #####################

BookingDataWithoutRed = filter(BookingData,BookingData$card=="red card")
##Eliminate matches with a red card in first 15 minutes
BookingDataWithout15 = filter(BookingDataWithoutRed, as.numeric(BookingDataWithoutRed$time) < 15)
BookingDataWithout15Red = filter(BookingData, !(BookingData$match_id %in% BookingDataWithout15$match_id))
##Either a match is going draw, and ended as draw. Or a match is being win by home, ended as home. Or a match is being win by away, ended as away. This is the perfect mathematics for this question. Other approaches might be hold, but there are too many possibilities, be careful!
MatchesDataWithoutSurprise = filter(MatchesData, abs(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score) + abs(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score - MatchesData$match_hometeam_extra_score + MatchesData$match_awayteam_extra_score) == 0 | (MatchesData$match_hometeam_extra_score - MatchesData$match_awayteam_extra_score)*(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score - MatchesData$match_hometeam_extra_score + MatchesData$match_awayteam_extra_score) > 0)

BetsDataWithoutSurprise1 = filter(BetsData1, BetsData1$match_id %in% MatchesDataWithoutSurprise$match_id)
BetsDataWithoutSurprise11 = filter(BetsDataWithoutSurprise1, BetsDataWithoutSurprise1$match_id %in% BookingDataWithout15Red$match_id)                                 

BetsData1HomeWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_1")
BetsData1AwayWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_2")
BetsData1DrawWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_x")

PHomeWin14 = 1/BetsData1HomeWithoutSurprise$value
PAwayWin14 = 1/BetsData1AwayWithoutSurprise$value
PDraw14 = 1/BetsData1DrawWithoutSurprise$value

plot(PHomeWin14-PAwayWin14, PDraw14,xlab= "P(Home)-P(Away)", ylab = "P(Draw)",col="red",main="BetVictor without Surprise")
TotalCountBin14=list()
CountBin14 = list()
MyProbability14 = list()
Bias14 = list()

for(i in 1:10){
  TotalCountBin14[i] = sum((PHomeWin14-PAwayWin14)>=(-1.2+0.2*i) & (PHomeWin14-PAwayWin14)<(-1.2+0.2*(i+1)))
}

BetsData1ActualDraw4 = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$match_id %in% DrawMatches$match_id)

BetsData1ActualDrawHome4 = filter(BetsData1ActualDraw4, BetsData1ActualDraw4$variable == "odd_1")
BetsData1ActualDrawAway4 = filter(BetsData1ActualDraw4, BetsData1ActualDraw4$variable == "odd_2")

PHomeWinActual14 = 1/BetsData1ActualDrawHome4$value
PAwayWinActual14 = 1/BetsData1ActualDrawAway4$value

for(i in 1:10){
  CountBin14[i] = sum((PHomeWinActual14-PAwayWinActual14)>=(-1.2+0.2*i) & (PHomeWinActual14-PAwayWinActual14)<(-1.2+0.2*(i+1)))
}

for(i in 1:10){
  MyProbability14[i] = CountBin14[[i]]/TotalCountBin14[[i]]
}

for(i in 1:10){
  Bias14[i] = sum((PHomeWin14-PAwayWin14)>(-1.2+0.2*i) & (PHomeWin14-PAwayWin14)<=(-1.2+0.2*(i+1)) & MyProbability14[i]>PDraw14 )
}
BiasArray14 = unlist(Bias14)
plot(x=1:10, y=BiasArray14,xlab="Index of the bin",ylab="Number of Biased Odds",main="BetVictor Without Surprise Bias Graph",xaxt = "n",yaxt="n")
axis(1, at = seq(1, 10, by = 1), las=2)
axis(2, at=seq(0,max(BiasArray1)+1,by=2), las=2)



#####Marathonbet#####


PremierMatches = filter(MatchesData, MatchesData$league_id == "148")

BetsData1 = filter(BetsData, BetsData$odd_bookmakers == "Marathonbet" & BetsData$match_id %in% PremierMatches$match_id )
BetsData1Home = filter(BetsData1, BetsData1$variable == "odd_1")
BetsData1Away = filter(BetsData1, BetsData1$variable == "odd_2")
BetsData1Draw = filter(BetsData1, BetsData1$variable == "odd_x")

###Part 1 Calculating the probabilities
PHomeWin1 = 1/BetsData1Home$value
PAwayWin1 = 1/BetsData1Away$value
PDraw1 = 1/BetsData1Draw$value

###Part 2 Normalization of probabilities
PHomeWin1Normal = PHomeWin1 / (PHomeWin1 + PAwayWin1 + PDraw1)
PAwayWin1Normal = PAwayWin1 / (PHomeWin1 + PAwayWin1 + PDraw1)
PDraw1Normal = PDraw1 / (PHomeWin1 + PAwayWin1 + PDraw1)

###Part 3
DrawMatches = filter(MatchesData, MatchesData$match_hometeam_score - MatchesData$match_awayteam_score == 0 & MatchesData$league_id == "148")
plot(PHomeWin1-PAwayWin1, PDraw1,xlab= "P(Home)-P(Away)", ylab = "P(Draw)",col="red",main="Marathonbet")
plot(PHomeWin1Normal-PAwayWin1Normal,PDraw1Normal,xlab= "P(Home)-P(Away) Normalized", ylab = "P(Draw) Normalized",col="blue",main="Marathonbet")
TotalCountBin1=list()
CountBin1 = list()
MyProbability1 = list()
Bias1 = list()
for(i in 1:10){
  TotalCountBin1[i] = sum((PHomeWin1-PAwayWin1)>=(-1.2+0.2*i) & (PHomeWin1-PAwayWin1)<(-1.2+0.2*(i+1)))
}

BetsData1ActualDraw = filter(BetsData1, BetsData1$match_id %in% DrawMatches$match_id)

BetsData1ActualDrawHome = filter(BetsData1ActualDraw, BetsData1ActualDraw$variable == "odd_1")

BetsData1ActualDrawAway = filter(BetsData1ActualDraw, BetsData1ActualDraw$variable == "odd_2")

PHomeWinActual1 = 1/BetsData1ActualDrawHome$value
PAwayWinActual1 = 1/BetsData1ActualDrawAway$value

for(i in 1:10){
  CountBin1[i] = sum((PHomeWinActual1-PAwayWinActual1)>=(-1.2+0.2*i) & (PHomeWinActual1-PAwayWinActual1)<(-1.2+0.2*(i+1)))
}

for(i in 1:10){
  MyProbability1[i] = CountBin1[[i]]/TotalCountBin1[[i]]
}

for(i in 1:10){
  Bias1[i] = sum((PHomeWin1-PAwayWin1)>(-1.2+0.2*i) & (PHomeWin1-PAwayWin1)<=(-1.2+0.2*(i+1)) & MyProbability1[i]>PDraw1 )
}
BiasArray1 = unlist(Bias1)
plot(x=1:10, y=BiasArray1,xlab="Index of the bin",ylab="Number of Biased Odds",main="Marathonbet Bias Graph",xaxt = "n",yaxt="n")
axis(1, at = seq(1, 10, by = 1), las=2)
axis(2, at=seq(0,max(BiasArray1)+1,by=5), las=2)

##If probability of two teams is similar, it is expected to have bias in the draw bet.

##################### Task 3 #####################

BookingDataWithoutRed = filter(BookingData,BookingData$card=="red card")
##Eliminate matches with a red card in first 15 minutes
BookingDataWithout15 = filter(BookingDataWithoutRed, as.numeric(BookingDataWithoutRed$time) < 15)
BookingDataWithout15Red = filter(BookingData, !(BookingData$match_id %in% BookingDataWithout15$match_id))
##Either a match is going draw, and ended as draw. Or a match is being win by home, ended as home. Or a match is being win by away, ended as away. This is the perfect mathematics for this question. Other approaches might be hold, but there are too many possibilities, be careful!
MatchesDataWithoutSurprise = filter(MatchesData, abs(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score) + abs(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score - MatchesData$match_hometeam_extra_score + MatchesData$match_awayteam_extra_score) == 0 | (MatchesData$match_hometeam_extra_score - MatchesData$match_awayteam_extra_score)*(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score - MatchesData$match_hometeam_extra_score + MatchesData$match_awayteam_extra_score) > 0)

BetsDataWithoutSurprise1 = filter(BetsData1, BetsData1$match_id %in% MatchesDataWithoutSurprise$match_id)
BetsDataWithoutSurprise11 = filter(BetsDataWithoutSurprise1, BetsDataWithoutSurprise1$match_id %in% BookingDataWithout15Red$match_id)                                 

BetsData1HomeWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_1")
BetsData1AwayWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_2")
BetsData1DrawWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_x")

PHomeWin14 = 1/BetsData1HomeWithoutSurprise$value
PAwayWin14 = 1/BetsData1AwayWithoutSurprise$value
PDraw14 = 1/BetsData1DrawWithoutSurprise$value

plot(PHomeWin14-PAwayWin14, PDraw14,xlab= "P(Home)-P(Away)", ylab = "P(Draw)",col="red",main="Marathonbet without Surprise")
TotalCountBin14=list()
CountBin14 = list()
MyProbability14 = list()
Bias14 = list()

for(i in 1:10){
  TotalCountBin14[i] = sum((PHomeWin14-PAwayWin14)>=(-1.2+0.2*i) & (PHomeWin14-PAwayWin14)<(-1.2+0.2*(i+1)))
}

BetsData1ActualDraw4 = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$match_id %in% DrawMatches$match_id)

BetsData1ActualDrawHome4 = filter(BetsData1ActualDraw4, BetsData1ActualDraw4$variable == "odd_1")
BetsData1ActualDrawAway4 = filter(BetsData1ActualDraw4, BetsData1ActualDraw4$variable == "odd_2")

PHomeWinActual14 = 1/BetsData1ActualDrawHome4$value
PAwayWinActual14 = 1/BetsData1ActualDrawAway4$value

for(i in 1:10){
  CountBin14[i] = sum((PHomeWinActual14-PAwayWinActual14)>=(-1.2+0.2*i) & (PHomeWinActual14-PAwayWinActual14)<(-1.2+0.2*(i+1)))
}

for(i in 1:10){
  MyProbability14[i] = CountBin14[[i]]/TotalCountBin14[[i]]
}

for(i in 1:10){
  Bias14[i] = sum((PHomeWin14-PAwayWin14)>(-1.2+0.2*i) & (PHomeWin14-PAwayWin14)<=(-1.2+0.2*(i+1)) & MyProbability14[i]>PDraw14 )
}
BiasArray14 = unlist(Bias14)
plot(x=1:10, y=BiasArray14,xlab="Index of the bin",ylab="Number of Biased Odds",main="Marathonbet Without Surprise Bias Graph",xaxt = "n",yaxt="n")
axis(1, at = seq(1, 10, by = 1), las=2)
axis(2, at=seq(0,max(BiasArray1)+1,by=2), las=2)


#####MARS BET#####


PremierMatches = filter(MatchesData, MatchesData$league_id == "148")

BetsData1 = filter(BetsData, BetsData$odd_bookmakers == "Marsbet" & BetsData$match_id %in% PremierMatches$match_id )
BetsData1Home = filter(BetsData1, BetsData1$variable == "odd_1")
BetsData1Away = filter(BetsData1, BetsData1$variable == "odd_2")
BetsData1Draw = filter(BetsData1, BetsData1$variable == "odd_x")

###Part 1 Calculating the probabilities
PHomeWin1 = 1/BetsData1Home$value
PAwayWin1 = 1/BetsData1Away$value
PDraw1 = 1/BetsData1Draw$value

###Part 2 Normalization of probabilities
PHomeWin1Normal = PHomeWin1 / (PHomeWin1 + PAwayWin1 + PDraw1)
PAwayWin1Normal = PAwayWin1 / (PHomeWin1 + PAwayWin1 + PDraw1)
PDraw1Normal = PDraw1 / (PHomeWin1 + PAwayWin1 + PDraw1)

###Part 3
DrawMatches = filter(MatchesData, MatchesData$match_hometeam_score - MatchesData$match_awayteam_score == 0 & MatchesData$league_id == "148")
plot(PHomeWin1-PAwayWin1, PDraw1,xlab= "P(Home)-P(Away)", ylab = "P(Draw)",col="red",main="Marsbet")
plot(PHomeWin1Normal-PAwayWin1Normal,PDraw1Normal,xlab= "P(Home)-P(Away) Normalized", ylab = "P(Draw) Normalized",col="blue",main="Marsbet")
TotalCountBin1=list()
CountBin1 = list()
MyProbability1 = list()
Bias1 = list()
for(i in 1:10){
  TotalCountBin1[i] = sum((PHomeWin1-PAwayWin1)>=(-1.2+0.2*i) & (PHomeWin1-PAwayWin1)<(-1.2+0.2*(i+1)))
}

BetsData1ActualDraw = filter(BetsData1, BetsData1$match_id %in% DrawMatches$match_id)

BetsData1ActualDrawHome = filter(BetsData1ActualDraw, BetsData1ActualDraw$variable == "odd_1")

BetsData1ActualDrawAway = filter(BetsData1ActualDraw, BetsData1ActualDraw$variable == "odd_2")

PHomeWinActual1 = 1/BetsData1ActualDrawHome$value
PAwayWinActual1 = 1/BetsData1ActualDrawAway$value

for(i in 1:10){
  CountBin1[i] = sum((PHomeWinActual1-PAwayWinActual1)>=(-1.2+0.2*i) & (PHomeWinActual1-PAwayWinActual1)<(-1.2+0.2*(i+1)))
}

for(i in 1:10){
  MyProbability1[i] = CountBin1[[i]]/TotalCountBin1[[i]]
}

for(i in 1:10){
  Bias1[i] = sum((PHomeWin1-PAwayWin1)>(-1.2+0.2*i) & (PHomeWin1-PAwayWin1)<=(-1.2+0.2*(i+1)) & MyProbability1[i]>PDraw1 )
}
BiasArray1 = unlist(Bias1)
plot(x=1:10, y=BiasArray1,xlab="Index of the bin",ylab="Number of Biased Odds",main="Marsbet Bias Graph",xaxt = "n",yaxt="n")
axis(1, at = seq(1, 10, by = 1), las=2)
axis(2, at=seq(0,max(BiasArray1)+1,by=5), las=2)

##If probability of two teams is similar, it is expected to have bias in the draw bet.

##################### Task 3 #####################

BookingDataWithoutRed = filter(BookingData,BookingData$card=="red card")
##Eliminate matches with a red card in first 15 minutes
BookingDataWithout15 = filter(BookingDataWithoutRed, as.numeric(BookingDataWithoutRed$time) < 15)
BookingDataWithout15Red = filter(BookingData, !(BookingData$match_id %in% BookingDataWithout15$match_id))
##Either a match is going draw, and ended as draw. Or a match is being win by home, ended as home. Or a match is being win by away, ended as away. This is the perfect mathematics for this question. Other approaches might be hold, but there are too many possibilities, be careful!
MatchesDataWithoutSurprise = filter(MatchesData, abs(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score) + abs(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score - MatchesData$match_hometeam_extra_score + MatchesData$match_awayteam_extra_score) == 0 | (MatchesData$match_hometeam_extra_score - MatchesData$match_awayteam_extra_score)*(MatchesData$match_hometeam_score - MatchesData$match_awayteam_score - MatchesData$match_hometeam_extra_score + MatchesData$match_awayteam_extra_score) > 0)

BetsDataWithoutSurprise1 = filter(BetsData1, BetsData1$match_id %in% MatchesDataWithoutSurprise$match_id)
BetsDataWithoutSurprise11 = filter(BetsDataWithoutSurprise1, BetsDataWithoutSurprise1$match_id %in% BookingDataWithout15Red$match_id)                                 

BetsData1HomeWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_1")
BetsData1AwayWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_2")
BetsData1DrawWithoutSurprise = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$variable == "odd_x")

PHomeWin14 = 1/BetsData1HomeWithoutSurprise$value
PAwayWin14 = 1/BetsData1AwayWithoutSurprise$value
PDraw14 = 1/BetsData1DrawWithoutSurprise$value

plot(PHomeWin14-PAwayWin14, PDraw14,xlab= "P(Home)-P(Away)", ylab = "P(Draw)",col="red",main="Marsbet without Surprise")
TotalCountBin14=list()
CountBin14 = list()
MyProbability14 = list()
Bias14 = list()

for(i in 1:10){
  TotalCountBin14[i] = sum((PHomeWin14-PAwayWin14)>=(-1.2+0.2*i) & (PHomeWin14-PAwayWin14)<(-1.2+0.2*(i+1)))
}

BetsData1ActualDraw4 = filter(BetsDataWithoutSurprise11, BetsDataWithoutSurprise11$match_id %in% DrawMatches$match_id)

BetsData1ActualDrawHome4 = filter(BetsData1ActualDraw4, BetsData1ActualDraw4$variable == "odd_1")
BetsData1ActualDrawAway4 = filter(BetsData1ActualDraw4, BetsData1ActualDraw4$variable == "odd_2")

PHomeWinActual14 = 1/BetsData1ActualDrawHome4$value
PAwayWinActual14 = 1/BetsData1ActualDrawAway4$value

for(i in 1:10){
  CountBin14[i] = sum((PHomeWinActual14-PAwayWinActual14)>=(-1.2+0.2*i) & (PHomeWinActual14-PAwayWinActual14)<(-1.2+0.2*(i+1)))
}

for(i in 1:10){
  MyProbability14[i] = CountBin14[[i]]/TotalCountBin14[[i]]
}

for(i in 1:10){
  Bias14[i] = sum((PHomeWin14-PAwayWin14)>(-1.2+0.2*i) & (PHomeWin14-PAwayWin14)<=(-1.2+0.2*(i+1)) & MyProbability14[i]>PDraw14 )
}
BiasArray14 = unlist(Bias14)
plot(x=1:10, y=BiasArray14,xlab="Index of the bin",ylab="Number of Biased Odds",main="Marsbet Without Surprise Bias Graph",xaxt = "n",yaxt="n")
axis(1, at = seq(1, 10, by = 1), las=2)
axis(2, at=seq(0,max(BiasArray1)+1,by=2), las=2)


```
